```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl

plt.style.use('science')

mpl.rcParams['figure.figsize'] = (15, 4)
mpl.rcParams['font.size'] = 16
mpl.rcParams['axes.grid'] = True
# # mpl.rcParams['axes.grid.axis']='y'
# mpl.rcParams['legend.markerscale']=8
# mpl.rcParams['text.usetex']=True

data=pd.read_csv('battery_dsoc_p.csv')
data=data[::100]
colors=['dimgray','red','blue']

plt.plot(data.power_t200_fil.values,data.dsoc.values,'.',color=colors[0],label=r'$\Delta s$ measurement',markersize=0.5,alpha=1)
plt.plot(data.power_t200_fil,data.dsoc_m4,'.',label='$\Delta s$ EPWL ',markersize=2,alpha=0.9)
plt.plot(data.power_t200_fil.values,data.dsoc_m1.values,'.',label='$\Delta s$ CC',markersize=2,alpha=0.9)
# plt.plot(data.power_t200_fil.values,data.dsoc_m5.values,'.',label='$\Delta s$ ANN',markersize=1)
plt.ylabel(r'$\Delta$SoC[\%]')
plt.ylim([-1,1])
plt.xlabel(r'$p_b$ battery power [kW]')
plt.xlim([-50,50])
plt.legend(markerscale=8)
plt.yticks(np.arange(-1,1.1,0.5))
plt.savefig('dSoC_P.pdf')
```
